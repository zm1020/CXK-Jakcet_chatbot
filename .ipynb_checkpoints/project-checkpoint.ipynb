# Jacket Recommender System — Project

This notebook demonstrates **important aspects** of our two CS175 jacket recommender pipelines on a Canada Goose–style catalog. It is designed to run in **under about 1 minute** on a GPU.

**What we show:**
- Import a sample of the data - We load and display a small sample from the product database used for training/retrieval.
- Import models we built/use — Pipeline 1 uses embedding models (sentence-transformers/all-MiniLM-L6-v2) for hybrid retrieval. Pipeline 2 uses a constraint parser and BM25 retrieval (and optionally FLAN-T5 from HuggingFace).
- Generate examples of predictions — We show examples where the model works well and examples where the model works poorly for both pipelines.

**Pipeline 1 (first):** Hybrid retrieval combining keyword embeddings and product-description semantic search.

**Pipeline 2 (second):** Constraint parsing (rule-based / FLAN-T5) plus BM25 retrieval.

```python
import sys
import os
from pathlib import Path

os.environ["TRANSFORMERS_NO_TF"] = "1"

REPO_ROOT = Path(".").resolve()
P1_ROOT = REPO_ROOT / "Pipeline 1" / "Jacket-chatbot"
P2_ROOT = REPO_ROOT / "Pipeline 2 Flan"
```

## Pipeline 1: Hybrid retrieval (keyword + semantic embeddings)

We import a **sample of the product data**, load the **embedding model from HuggingFace** (`sentence-transformers/all-MiniLM-L6-v2`), and generate **examples of predictions** — where the model works well and where it works poorly.

### 1.1 Load a sample of the data

We import a small sample from the Canada Goose product database (`data/canada_goose.db`).

```python
P1_SRC = P1_ROOT / "src"
if str(P1_SRC) not in sys.path:
    sys.path.insert(0, str(P1_SRC))
os.chdir(P1_ROOT)

import sqlite3
import pandas as pd
conn_p1 = sqlite3.connect("data/canada_goose.db")
conn_p1.row_factory = sqlite3.Row

products_p1 = pd.read_sql_query(
    "SELECT id, name, gender, price, currency FROM products LIMIT 8",
    conn_p1,
)
print("Sample of the product data:")
display(products_p1)
print("\nFull catalog: 96 products in data/canada_goose.db")
```

### 1.2 Load model(s) from HuggingFace

We load the embedding model **from HuggingFace** (`sentence-transformers/all-MiniLM-L6-v2`) and build the keyword embedder and product-description index used for hybrid retrieval.

```python
from chatbot.embedder import KeywordEmbedder, ProductDescriptionEmbedder, DOMAIN_KEYWORDS
from chatbot.chatbot_runner import retrieve_and_rank_hybrid, ScoredProduct

emb = KeywordEmbedder(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    cache_dir="data/embeddings",
    keywords=DOMAIN_KEYWORDS,
)
desc_index = ProductDescriptionEmbedder(data_dir="data")
desc_index.ensure_loaded(conn_p1)
print("Models loaded from HuggingFace (sentence-transformers/all-MiniLM-L6-v2).")
```

### 1.3 Examples where the model works well

Queries that mention warmth, product type (jacket), use case (winter, travel), or budget are matched well by the hybrid (keyword + semantic) retrieval.

```python
def show_p1(query, price_max=None, gender=None):
    r, kw = retrieve_and_rank_hybrid(
        conn_p1, emb, desc_index, query,
        price_max=price_max, gender=gender, return_k=5,
    )
    print(f'Query: "{query}"')
    print(f"Matched keywords: {[k for k, _ in kw[:6]]}\n")
    display(pd.DataFrame([{"name": p.name, "price": p.price, "score": round(p.score, 3)} for p in r]))

show_p1("warm down jacket for winter under $1000", price_max=1000.0)
show_p1("women's travel jacket", price_max=1000.0, gender="women")
```

### 1.4 Examples where the model works poorly

Vague queries ("something nice") or out-of-vocabulary terms (e.g. "red", "coat" when our vocab favors "jacket"/"parka") can yield generic or less relevant rankings.

```python
show_p1("something nice")
show_p1("red coat under $500", price_max=500.0)

conn_p1.close()
os.chdir(REPO_ROOT)
```

## Pipeline 2: Constraint parsing + BM25/FLAN retrieval

We use the Pipeline 2 codebase: rule-based (or FLAN) query parsing into structured constraints, then BM25 retrieval over the product catalog.

```python
ROOT = P2_ROOT
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

from src.pipeline import JacketRecommenderSystem
from src.db import connect_db, load_product_tables
import pandas as pd
```

### 2.1 Load a sample of the data

We use the Canada Goose product database (`canada_goose.db`).

```python
DB_PATH = ROOT / "canada_goose.db"
conn = connect_db(str(DB_PATH))
products_df, keywords_df = load_product_tables(conn)

print("Products sample (columns:", list(products_df.columns), ")")
display(products_df.head(8))

print("\nProduct keywords sample (columns:", list(keywords_df.columns), ")")
display(keywords_df.head(10))

print(f"\nDataset size: {len(products_df)} products, {len(keywords_df)} keyword rows")
conn.close()
```

### 2.2 Building our model

We instantiate the pipeline. The system preprocesses products, builds a retriever, and parses queries into constraints.

```python
system = JacketRecommenderSystem(db_path=str(DB_PATH))
print("Pipeline ready. Summary stats:")
import json
print(json.dumps(system.summary_stats(), indent=2))
```

### 2.3 Examples where the model works well

Queries that mention product type, price (e.g. "under $X"), use case (travel, hiking, extreme cold), or gender are parsed reliably and return relevant results.

```python
def show_result(query: str, method: str = "baseline", top_k: int = 5):
    result, ranked = system.query(query, method=method, top_k=top_k)
    print(f"Query: \"{query}\"")
    print(f"Parsed constraints: {result['constraints']}")
    print(f"Runtime: {result['runtime_sec']}s | Results: {result['num_results']}\n")
    if not ranked.empty:
        cols = [c for c in ["name", "price", "product_type", "score"] if c in ranked.columns]
        display(ranked[cols].head(top_k))
    else:
        print("No results.")
    print("---")

# Example 1: Clear type + price + use case
show_result("very warm long down jacket for extreme cold under $1500")

# Example 2: Gender + type + price
show_result("women's travel jacket under $1000")

# Example 3: Type + use case (no price)
show_result("packable jacket for travel")
```

### 2.4 Examples where the model works poorly

- Very vague queries (e.g. "something warm") may yield generic or irrelevant rankings.
- Out-of-vocabulary product types (e.g. "coat" when we only have jacket/parka/vest) may not map correctly.
- Contradictory or unusual constraints can produce empty or odd results.

```python
# Vague: few constraints extracted, results may be generic
show_result("something warm for winter")

# "Coat" is not in our type vocabulary (we have jacket, parka, vest, hoody, sweater)
# so product_type may be missing or wrong
show_result("women's coat under $800")
```
